```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
one_hot = pd.read_csv(r"/Users/zacksisco/Documents/BoulderCleaned.csv")
one_hot.head()
```

```python
one_hot.shape
```

```python
def price_class(x): 
    if (x >= 80000 and x <= 264000):
        return 1
    if (x >= 264001 and x <= 448000):
        return 2
    if (x >= 448001 and x <= 632000):
        return 3
    if (x >= 632001 and x <= 816000):
        return 4
    if (x >= 816001 and x <= 1000000):
        return 5
    if x >= 1000001: 
        return 6
```

```python
one_hot_copy = one_hot.copy()
one_hot_copy.drop('ZIPCODE', axis=1, inplace=True)
one_hot_copy.head()
```

```python
one_hot_copy['HOUSE_CLASS'] = one_hot_copy['SALE_PRICE'].apply(price_class)
one_hot_copy.head()
```

```python
onehotnorm1 = one_hot_copy.copy()
onehotnorm1.head()
```

```python
onehotnorm = onehotnorm1.copy()
onehotnorm
```

```python
onehotnorm.drop('HOUSE_CLASS', axis=1, inplace=True)
onehotnorm.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.dummy import DummyClassifier
from sklearn.model_selection import KFold
knn = KNeighborsClassifier(n_neighbors=5)
dummy = DummyClassifier(strategy='uniform')
```

# Dummy Classification

```python
nx2 = pt_one_hot_norm.drop(['HOUSE_CLASS'] , axis=1)
ny2 = pt_one_hot_norm['HOUSE_CLASS']
print(nx2.shape, ny2.shape)

X2 = nx2
Y2 = ny2
kf2 = KFold(n_splits = 10)
kf2.get_n_splits(X2)
X2.head()
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score, confusion_matrix
import seaborn as sns

class_metrics = {'accuracy':[], 'f1':[], 'precision':[], 'recall':[]}

for train_index, test_index in kf2.split(X2):
    print("TRAIN:", train_index, "TEST:", test_index, '\n')
    X2_train, X2_test = X2.iloc[train_index], X2.iloc[test_index]
    Y2_train, Y2_test = Y2.iloc[train_index], Y2.iloc[test_index]
    
    dummy.fit(X2_train, Y2_train)
    #knn.fit(X2_train, Y2_train)
    
    dummy_pred = dummy.predict(X2_test)
    #knn_pred = knn.predict(X2_test)
    
    print('Dummy Accuracy Score: ', dummy.score(X2_train, Y2_train))
    print()
    print('Dummy F1 Score: ', f1_score(Y2_test, dummy_pred, average='weighted'))
    print()
    print('Dummy Precision: ', precision_score(Y2_test, dummy_pred, average='weighted'))
    print()
    print('Dummy Recall: ', recall_score(Y2_test, dummy_pred, average='weighted'))
    print()
    
    class_metrics['accuracy'].append(dummy.score(X2_train, Y2_train))
    class_metrics['f1'].append(f1_score(Y2_test, dummy_pred, average='weighted'))
    class_metrics['precision'].append(precision_score(Y2_test, dummy_pred, average='weighted'))
    class_metrics['recall'].append(recall_score(Y2_test, dummy_pred, average='weighted'))
```

```python
cf_matrix = confusion_matrix(Y2_test, dummy_pred)
sns.heatmap(cf_matrix, annot=True, cmap='Blues')
    
```

```python
dummy_metrics = pd.DataFrame(data=class_metrics)
dummy_metrics
```

```python
dummys_metrics = dummy_metrics.mean(axis=0)
dummys_metrics
```

```python
dummy_metrics = pd.DataFrame(data=dummys_metrics, columns=['Dummy Classification'])
dummy_metrics = dummy_metrics.transpose()
dummy_metrics
```
